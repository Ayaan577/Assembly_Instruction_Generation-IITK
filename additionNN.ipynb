```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Data Preparation
np.random.seed(42)
tf.random.set_seed(42)

# Generate data
num_samples = 1000
max_value = 100
X = np.random.randint(0, max_value, size=(num_samples, 2)).astype(np.float64)  # Ensure data is float64
y = (X[:, 0] + X[:, 1]).astype(np.float64)

# Normalize the data
X_normalized = X / max_value
y_normalized = y / max_value

# Model Definition
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,), dtype=tf.float64),  # Input layer with float64
    tf.keras.layers.Dense(10, activation='relu', dtype=tf.float64),
    tf.keras.layers.Dense(10, activation='relu', dtype=tf.float64),
    tf.keras.layers.Dense(1, dtype=tf.float64)  # Output layer with float64
])

model.compile(optimizer='adam', loss='mean_squared_error')

# Training the Model
model.fit(X_normalized, y_normalized, epochs=200, batch_size=32, verbose=1)

# Testing the Model
test_data = np.array([[10, 15], [20, 25], [5, 7]], dtype=np.float64)
test_data_normalized = test_data / max_value
predictions = model.predict(test_data_normalized)

# Denormalize predictions
predictions_denormalized = predictions.flatten() * max_value

# Results
for i, (input_vals, pred) in enumerate(zip(test_data, predictions_denormalized)):
    actual = input_vals.sum()
    print(f"Input: {input_vals}, Predicted Sum: {pred:.2f}, Actual Sum: {actual}")
```
